### EMPLOYEES 테이블에서 JOB_ID가 it_prog인 사원의 이름과 급여를 출력하세요
#### 조건1. 비교하기 위한 값은 소문자로 입력해야 함(lower 이용)
#### 조건2. 이름은 앞 3문자까지 출력, 나머지는 *로 출력, 이 열의 별칭은 name(rpad, substr, length 이용)
#### 조건3. 급여는 전체 10자리로 출력하되, 나머지는 *로 출력, 이 열의 별칭은 salary(lpad 이용)

```python
%%sql
SELECT  RPAD(SUBSTR(first_name, 1,3), length(first_name),'*') AS name, RPAD(salary, 10, '*') AS salary
FROM employees
WHERE LOWER(job_id) = 'it_prog'
```

```python
%%sql
SELECT SUBSTR(first_name, 1,3)
FROM employees
WHERE LOWER(job_id) = 'it_prog'
```

```python
%%sql
SELECT col1, REGEXP_INSTR(col1, '[A-B]') AS data1,
REGEXP_INSTR(col1, '%') AS data2
FROM test_regexp
--REGEXP_INSTR을 이용하여 (시작하는) 특정 문자열의 위치를 출력
```

```python
%%sql
SELECT
ROUND(SUM(salary*commission_pct),2) AS sum_bonus,
COUNT(*) AS count,
ROUND(AVG(salary*commission_pct),2) AS avg_bonus,
ROUND(SUM(salary*commission_pct)/count(*),2) AS avg_bonus2
FROM employees
--NULL처리 잘생각하기, 첫번째 보너스는 보너스
```

```python
%%sql
SELECT department_id, AVG(salary)
FROM employees
GROUP BY department_id
    
--그룹함수 사용할 때에는 NULL처리가 어떻게 되냐에 따라 값이 달라진다
```

```python
%%sql
SELECT department_id, job_id, AVG(salary)
FROM employees
GROUP BY department_id, job_id
HAVING AVG(salary) > 8000
--'WHERE AVG(salary) > 8000' where절을 바로 작성하여 이용할 수 없음
```
